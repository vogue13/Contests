```python
df['ClientOrder'] = df['ClientID']+df['Global Client Order Rank']/10
```

```python
df
```

```python
drops = df.groupby('ClientOrder')['Absence'].sum().sort_values(ascending=False).reset_index()
```

```python
drops
```

```python
drops[drops['Absenсe']>6]
```

```python
df['AbsenceCum'] = (df['Absence'] +df['AbsenceCum'].shift(-1))*df['Absence']
```

```python
df['AbsenceCum']=  df[df['Service Usage Sign'] == 0].sort_values(['ClientOrder','Date']).groupby('ClientOrder')['Absence'].cumsum()
```
